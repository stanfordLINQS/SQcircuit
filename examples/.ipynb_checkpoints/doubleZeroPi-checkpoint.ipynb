```python
loop1 = sq.Loop()
loop2 = sq.Loop()

CJ = sq.Capacitor(10, "GHz")
JJ1 = sq.Junction(5, "GHz", loops=[loop1], cap = CJ)
JJ2 = sq.Junction(5, "GHz", loops=[loop2], cap = CJ)
L = sq.Inductor(5, "GHz", loops=[loop1,loop2])

circuitElemetns = {(0, 1): [JJ1],
                   (1, 2): [JJ1],
                   (0, 3): [JJ2],
                   (2, 3): [JJ2],
                   (0, 2): [L]}
```

```python
def getCmap(cmap):
    # get the transparent colormap
    

    ncolors = 256
    color_array = plt.get_cmap(cmap)(range(ncolors))
    color_array[:,-1] = np.linspace(0,1,ncolors)
    map_object = LinearSegmentedColormap.from_list(name='Transparent',colors=color_array)
    return map_object
```

```python
from circuitClass import *
# the value of the capacitors, inductors and JJ Energies.
E_C = 0.15*4
E_J = 5/(2*np.pi*hbar)
E_L = 0.13
C = e**2/2/E_C
L = (Phi0/2/np.pi)**2/E_L
Cshunt = C
Cx = C*0
```

```python
circuitParam={'graph':[[0,1],[1,2],[2,3],[3,0],[2,0]],
            'inductors':[None,None,None,None,L],
            'capacitors':[C,C,C,C,0],
            'JJs':[E_J,E_J,E_J,E_J,None]}
cr1 = Qcircuit(circuitParam)
```

```python
cr1.setModeNumbers([27,17,17])
cr1.configure()
```
